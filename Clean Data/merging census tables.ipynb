```python
import pandas as pd
```

```python
data_load1 = "2016 educational attainment.csv"
data_read1 = pd.read_csv(data_load1)
data_load2 = "2016 proverty_unemployment_income.csv"
data_read2 = pd.read_csv(data_load2)
```

```python
#merge the 2 tables
merged_census = pd.merge(data_read1, data_read2, on = "Zip Code", how = "outer")
merged_census
```
